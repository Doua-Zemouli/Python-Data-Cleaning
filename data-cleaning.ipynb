# Importing library and data

```python
import pandas as pd
```

```python
df = pd.read_excel('/home/dell/Téléchargements/Data/jnotebook/Customer Call List.xlsx')
df.head()
```

# Starting the cleaning

```python
# Dropping duplicates
df = df.drop_duplicates()
```

```python
# Dropping useless columns
df = df.drop(columns = 'Not_Useful_Column')
```

```python
# Removing typos
df["Last_Name"] = df["Last_Name"].str.strip("./_")
```

```python
# Standardize columns values
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','')
df["Address"] = df["Address"].str.replace('N/a','')
```

```python
#
df[["Street", "State", "zip_code"]] = df["Address"].str.split(',',expand=True)
```

```python
#
df = df.drop(columns = 'zip_code')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
```

```python
# Finally filling any missing value with void
df = df.fillna('')
```

```python
# Dropping customers that don't want to be called
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
```

```python
# Dropping customers that didn't provide a phone number
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
```

```python
# Fixing the index
df = df.reset_index(drop=True)
```

# Export the DataFrame to an Excel file

```python
df.to_excel('call-list-cleaned.xlsx', index=False)
print("\nData has been exported to call-list-cleaned.xlsx")
```
